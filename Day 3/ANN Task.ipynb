### Implementing Gradient Descent From Scratch

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def log_loss(y_actual, y_predicted):
    total_error = 0
    for y, y_hat in zip(y_actual, y_predicted):
        factor = 1e-15
        match(y_hat):
            case 0:
                y_hat += factor
            case 1:
                y_hat -= factor
            case _:
                pass
        total_error += y * np.log(y_hat) + (1-y) * np.log(1-y_hat)
    print("Total Error:",total_error)
    return -(total_error / len(y_actual))
```

```python
def gradient_descent(age, aff, y_actual, epochs):
    w1 = 1
    w2 = 1
    bias = 0
    learning_rate = 0.5
    n = len(y_actual)

    for i in range(epochs):
        weighted_sum = w1 * age + w2 * aff + bias
        y_pred = sigmoid(weighted_sum)
        loss = log_loss(y_actual, y_pred)
        
        w1_d = (1/n) * np.dot(age, (y_pred - y_actual))
        w2_d = (1/n) * np.dot(aff, (y_pred - y_actual))
        bias_d = np.mean(y_pred - y_actual)
        
        w1 = w1 - learning_rate * w1_d
        w2 = w2 - learning_rate * w2_d
        bias = bias - learning_rate * bias_d

        print("Epoch",i,"w1",w1,"w2",w2,"bias",bias,"loss",loss)
    return w1,w2,bias
```

### Implementing ANN from Scratch

```python
class ANN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
        
    def fit(self,X,y,epochs):
        self.gradient_descent(X["age"],X["affordibility"] , y , epochs)

    def gradient_descent(self, age, aff, y_actual, epochs):   
        learning_rate = 0.5
        n = len(y_actual)
    
        for i in range(epochs):
            weighted_sum = self.w1 * age + self.w2 * aff + self.bias
            y_pred = sigmoid(weighted_sum)
            loss = log_loss(y_actual, y_pred)
            
            w1_d = (1/n) * np.dot(age, (y_pred - y_actual))
            w2_d = (1/n) * np.dot(aff, (y_pred - y_actual))
            bias_d = np.mean(y_pred - y_actual)
            
            self.w1 = self.w1 - learning_rate * w1_d
            self.w2 = self.w2 - learning_rate * w2_d
            self.bias = self.bias - learning_rate * bias_d
    
            print("Epoch",i,"w1",self.w1,"w2",self.w2,"bias",self.bias,"loss",loss)
        return self.w1, self.w2, self.bias
```
